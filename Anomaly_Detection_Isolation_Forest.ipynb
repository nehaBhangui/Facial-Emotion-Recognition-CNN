```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
# Normal data (majority)
normal_data = np.random.randn(300, 2)

# Anomalies (few + far away)
anomalies = np.random.uniform(low=-6, high=6, size=(20, 2))

# Combine them
X = np.vstack((normal_data, anomalies))
```

```python
plt.scatter(X[:, 0], X[:, 1])
plt.title("Data BEFORE Anomaly Detection")
plt.show()
```

```python
from sklearn.ensemble import IsolationForest
```

```python
model = IsolationForest(
    n_estimators=100,
    contamination=0.06,
    random_state=42
)

model.fit(X)
```

```python
predictions = model.predict(X)
```

```python
normal = X[predictions == 1]
outliers = X[predictions == -1]
```

```python
plt.scatter(normal[:, 0], normal[:, 1], label="Normal")
plt.scatter(outliers[:, 0], outliers[:, 1], color="red", label="Anomaly")
plt.legend()
plt.title("Detected Anomalies")
plt.show()
```
